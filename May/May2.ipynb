
**On an average how many hectares of area was destroyed by the fires in the month of August?**



```python
data.query("month == 'aug'")['area'].mean()
```

**In the month of August, what was the week-day, when the area destroyed in the fires was maximum but the temperature was least**.


```python
data.query("month == 'aug' and day not in ('sat','sun')").sort_values(by = ['area','temp'], ascending =[False,True])
```

**In the month of August, what is the maximum value of ‘rain’ in the 7th decile of wind speed column**

```python
data['wind'].quantile(0.7)
```

```python
data.query("wind>=4.5 or wind<=4.9 and month=='aug'")['rain'].max()
```

**Maximum rainfall happened on:**


```python
data.groupby(["day"]).agg({'rain':'max'})
```
