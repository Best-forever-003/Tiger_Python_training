```python
mag=[]
loc=[]
lat=[]
lon=[]

for row in json_data['features']:
  mag.append(row['properties']['mag'])
  loc.append(row['properties']['place'])
  lat.append(row['geometry']['coordinates'][1])
  lon.append(row['geometry']['coordinates'][2])
```

```python
l=len(json_data)
mag=[]
coor=[]
for i in range(l):
    mag.append(json_data['features'][i]['properties']['mag'])
    coor.append(json_data['features'][i]['geometry']['coordinates'])
print(mag)
print(coor)
```
