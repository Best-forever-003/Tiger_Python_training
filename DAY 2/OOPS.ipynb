```python
class Account():
  def __init__(self,balance):
    self.balance=balance

  def withdraw(self,amount):
    self.balance=self.balance - amount
    print(f"New Balance is {self.balance} and amount withdrawn is {amount}")

  def deposit(self,amount):
    self.balance=self.balance + amount
    print(f"New Balance is {self.balance} and amount deposited is {amount}")
```
